```python
# a)

import requests
import pandas as pd

def fetch(data_source, signal):
    
    #Query Data
    response = requests.get("https://api.delphi.cmu.edu/epidata/covidcast/", params = {
            "data_source": data_source,
            "signal": signal,
            "time": 'day:20220101',
            "geo": 'state:*'
        })

    #convert to json
    result = response.json()
    
    #convert to Pandas DataFrame
    results = pd.DataFrame(result['epidata'])
    
    return results
    
    

scl_fb = fetch('fb-survey', 'smoothed_wcli')
shccl_fb = fetch('fb-survey', 'smoothed_whh_cmnty_cli')
conf_jhu = fetch('jhu-csse', 'confirmed_7dav_incidence_prop')

#conf_jhu.info()

#b)

#create table
table = pd.DataFrame()

#add columns
table['smoothed_wcli'] = scl_fb['value']
table['smoothed_whh_cmnty_cli'] = shccl_fb['value']
table['confirmed_7dav_incidence_prop'] = conf_jhu['value']

#compute correlation matrix
correlation_matrix = table.corr()
print(correlation_matrix)

```
